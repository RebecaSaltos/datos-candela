Para cada muestra de sangre se midió tal proteína en ambos laboratorios y las diferencias entre laboratorios (AB) fueron las siguientes: 0.6, 0.7, 0.8, 0.9, 0.3, 0.5, -0.5, 1.3, 0.4, 0.8 Considerando normalidad, calcular un intervalo  de confianza para la diferencia media de medición al nivel de confianza 0.9 considerando que la desviación típica de las diferencias poblacionales es 0.22.  A este nivel de confianza, qué tamaño mínimo de muestra deberíamos tomar para que la amplitud del intervalo fue menor o igual que la mitad del anterior?
